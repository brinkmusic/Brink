```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
from datetime import datetime
```

```python
# Authenticate -- opens a browser window on first run to grant permission
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id=CLIENT_ID,
    client_secret=CLIENT_SECRET,
    redirect_uri=REDIRECT_URI,
    scope="user-read-recently-played"
))
```

```python
# Fetch up to 50 recently played tracks (Spotify max per request)
results = sp.current_user_recently_played(limit=50)

rows = []
for item in results["items"]:
    track = item["track"]
    played_at = datetime.strptime(item["played_at"], "%Y-%m-%dT%H:%M:%S.%fZ")
    rows.append({
        "played_at":   played_at.strftime("%Y-%m-%d %H:%M"),
        "track":       track["name"],
        "artist":      ", ".join(a["name"] for a in track["artists"]),
        "album":       track["album"]["name"],
        "duration_s":  track["duration_ms"] // 1000,
        "spotify_url": track["external_urls"]["spotify"],
    })

df = pd.DataFrame(rows)
df
```

```python
# Quick summary
print(f"Total plays fetched : {len(df)}")
print(f"Unique tracks       : {df['track'].nunique()}")
print(f"Unique artists      : {df['artist'].nunique()}")
print()
print("Most played artists in this window:")
print(df["artist"].value_counts().head(10).to_string())
```
